Os campos da tabela incluem:

* **Item**: Identificação do registro.
* **Tipo**: Classificação atribuída ao registro.
  * **Ocorrência**: Registros relacionados a eventos que causaram desligamentos na subestação.
  * **Manobra**: Registros que não estão diretamente ligados a eventos, mas que podem ser consultados pelo operador para auxiliá-lo durante uma operação.
* **Eventos**: Registros de eventos que resultaram em uma ocorrência de desligamento.
* **Nome**: Designação do registro; quando o tipo é ocorrência, apresenta a identificação da ocorrência.
* **Recomposição**: Passos de manobra que o operador deve seguir para realizar a manobra solicitada ou restabelecer uma ocorrência.

O sistema de supervisão e controle deve interagir com o algoritmo. Ele será responsável por repassar todos os eventos ocorridos na subestação.
Estes eventos serão analisados pelo Gemini para fornecer o **Nome** e a manobra necessária para recompor o sistema (**Recomposição**).

*É importante destacar que a lógica necessária para esta integração não está dentro do escopo desta atividade.*

```python
model = "models/embedding-001"
```

```python
def embed_fn(model, title, text):
    embeddings = genai.embed_content(model=model,
                                 content=text,
                                 title=title,
                                 task_type="RETRIEVAL_DOCUMENT")
    return embeddings["embedding"]
```

```python
df["Embeddings"] = df.apply(lambda row: embed_fn("models/embedding-001", row["Nome"], row["Eventos"]), axis=1)
```

```python
def gerar_e_buscar_consulta(consulta, base, model):
    embedding_da_consulta = genai.embed_content(model=model,
                                 content=consulta,
                                 task_type="RETRIEVAL_QUERY")["embedding"]

    produtos_escalares = np.dot(np.stack(df["Embeddings"]), embedding_da_consulta)

    indice = np.argmax(produtos_escalares)
    return df.iloc[indice]["Nome"],df.iloc[indice]["Recomposicao"]
```

```python
generation_config = {
    "candidate_count": 1,
    "temperature": 0.5
}

model_2 = genai.GenerativeModel("gemini-1.0-pro", generation_config=generation_config)
```
